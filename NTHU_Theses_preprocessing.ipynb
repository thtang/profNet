```python
from collections import defaultdict

prof_teach_count = defaultdict(int)
prof_oral_count = defaultdict(lambda : defaultdict(float))

for prof, orals in zip(theses_df["指導教授"],theses_df["口試委員"]):
    prof = prof.split("\n")[0]
    prof_teach_count[prof] +=1

print(prof_teach_count)

for prof, orals in zip(theses_df["指導教授"],theses_df["口試委員"]):
    prof = prof.split("\n")[0]
    oral_list = orals.split("\n")
    for oral in oral_list:
        if oral != "":
            prof_oral_count[prof][oral]+= 1.0/prof_teach_count[prof]
```

```python
import operator
sorted(prof_oral_count["簡禎富"].items(), key=operator.itemgetter(1),reverse=True)
```

```python
prof_oral_pair_df = pd.DataFrame.from_dict({(i,j): prof_oral_count[i][j] 
                           for i in prof_oral_count.keys() 
                           for j in prof_oral_count[i].keys()},
                       orient='index')
prof_oral_pair_df.columns = ["prob"]
prof_oral_pair_df.head(3)
```

## Create word count for word cloud

```python
theses_df["指導教授_cleaned"] = theses_df["指導教授"].map(lambda x: x.split("\n")[0])
```

```python
theses_df["指導教授_cleaned"].head()
```

```python
from nltk.probability import FreqDist
prof_abstract = defaultdict(str)
for i in theses_df.iterrows():
    prof_abstract[i[1][24]] += i[1][25]

stop_word_list = open("stop_words.txt",encoding="utf8").read().split("\n")
stop_word_list = stop_word_list + ["針對","”","「","’","/","」","內","（","：","）","不","台","；"]
prof_word_count_dict = defaultdict(list)

for prof, abstract in prof_abstract.items():
    word_count_list = []
    fq_dict = FreqDist(abstract.split())
    for word, count in fq_dict.items():
        if word not in stop_word_list:
            word_count_list.append([word,count])
    prof_word_count_dict[prof] = word_count_list
```

```python
nodes =[]
for item in prof_color.items():
    nodes.append({"id":item[0], "group":item[1],"department":prof_department[item[0]],"wordcount":prof_word_count_dict[item[0]]})
```

```python
import json
with open('nodes.txt', 'w') as outfile:
    json.dump(nodes, outfile)
```
